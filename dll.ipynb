```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
class DoubleLinkedList:
    def __init__(self):
        self.head=None
    def InsertAtBegin(self,newdata):
        new_node=Node(newdata)

        if self.head is None:
            self.head=new_node
            return
        new_node.next=self.head
        self.head.prev=new_node
        self.head=new_node
    def displayforward(self):
        temp=self.head
        while temp:
            print(temp.data,end=" ")
            temp=temp.next
    def displaybackward(self):
        temp=self.head
        if temp is None:
            return
        while temp.next:
            temp=temp.next
        while temp:
            print(temp.data,end=" ")
            temp=temp.prev
        print("None")



DLL=DoubleLinkedList()
DLL.InsertAtBegin(10)
DLL.InsertAtBegin(20)
DLL.InsertAtBegin(30)
DLL.InsertAtBegin(40)
DLL.displayforward()
print()
DLL.displaybackward()
```
