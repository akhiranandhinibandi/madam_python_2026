```python
import heapq
heap=[5,7,9,1,3]
heapq.heapify(heap)
print("Heap:",heap)

heapq.heappush(heap,4)
print("Heap after push:",heap)
smallest=heapq.heappop(heap)

print(smallest)
print(heapq.heappop(heap))
```

```python
## one way
from collections import Counter
import heapq
def topKElements(arr,k):
  if k==len(arr):
    return arr
  count=Counter(arr)
  return heapq.nlargest(k,count.keys(),key=count.get)
arr=[1,1,1,2,1,2,3]
k=2
result=topKElements(arr,k)
print(result)
```

```python
import heapq
values=[5,1,3,7,4,2]
heapq.heapify(values)
print("Heap:",values)
n_smallest=heapq.nsmallest(3,values)
print("smallest 3 elements:",n_smallest)
n_largest=heapq.nlargest(2,values)
print("largest 2 elements:",n_largest)
```

```python
# k closest points from origin
from heapq import heappush,heappop
from math import sqrt
def get_distance(x,y):
  return sqrt(x**2+y**2)
def k_closest(points,k):
  min_heap=[]
  for x,y in points:
    heappush(min_heap,(get_distance(x,y),[x,y]))
  result=[]
  for i in range(k):
    result.append(heappop(min_heap))
  return result
points=[[3,3],[5,-1],[-2,4]]
k=3
result=k_closest(points,k)
print(result)
```
